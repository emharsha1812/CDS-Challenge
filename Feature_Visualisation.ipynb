```python
import pandas as pd
import numpy as np 
import seaborn as sns
from matplotlib import pyplot as plt

```

```python
data=pd.read_csv('FinalData.csv')
```

```python
# Count the number of movies for each director
movie_count_by_director = data['director_name'].value_counts().reset_index()
movie_count_by_director.columns = ['director_name', 'movie_count']

# Visualize the distribution of movie counts for directors
plt.figure(figsize=(12, 6))
sns.histplot(data=movie_count_by_director, x='movie_count', kde=True)
plt.title('Distribution of Movie Counts per Director')
plt.xlabel('Number of Movies')
plt.ylabel('Count')
plt.show()


#A Normal right skewed distirubtion shows that it is a useful feature
```

```python
#Distribution of Movie Durations
plt.figure(figsize=(10, 6))
sns.histplot(data['duration'], kde=True)
plt.title('Distribution of Movie Durations')
plt.xlabel('Duration (minutes)')
plt.ylabel('Count')
plt.show()

#A Normal right skewed distirubtion shows that it is a useful feature
```

```python
#Director Facebook Likes 
plt.figure(figsize=(10, 6))
sns.scatterplot(x='director_facebook_likes', y='imdb_score', data=data)
plt.title('Director Facebook Likes vs. IMDb Score')
plt.xlabel('Director Facebook Likes')
plt.ylabel('IMDb Score')
plt.show()

#Conclusion - Where Facebook likes are more, the IMDB Score is more
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='num_voted_users', y='imdb_score', data=data)
plt.title('Number of Voted Users vs. IMDb Score')
plt.xlabel('Number of Voted Users')
plt.ylabel('IMDb Score')
plt.show()

#Conclusion - when more people vote, the IMDB Score gets high
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='cast_total_facebook_likes', y='imdb_score', data=data)
plt.title('Total Cast Facebook Likes vs. IMDb Score')
plt.xlabel('Total Cast Facebook Likes')
plt.ylabel('IMDb Score')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['year_since_last_movie'], kde=True)
plt.title('Distribution of Years Since Last Movie')
plt.xlabel('Years Since Last Movie')
plt.ylabel('Count')
plt.show()

#Conclusion - Since this follows a normal distribution, it would be beneficial to include it as a feature
```
